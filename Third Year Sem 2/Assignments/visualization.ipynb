```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
QuickSortDF = pd.DataFrame({'Time':[41,43,43,68,129]})
MergeSortDF = pd.DataFrame({'Time':[61,49,55,100,156]})
elements = [50,100,1000,5000,10000]
```

```python
plt.title("QuickSort vs MergeSort")
plt.plot(elements,QuickSortDF.Time)
plt.plot(elements,MergeSortDF.Time)
plt.xlabel("Number of elements")
plt.ylabel("Time in ms")
plt.legend(["Quick Sort", "Merge Sort"])
plt.show()
```
